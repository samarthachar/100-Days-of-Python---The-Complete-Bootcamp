#  Pandas Basics  Learning Summary

This README captures key concepts and commands for working with pandas DataFrames in Python. It's a quick reference for exploring, cleaning, and analyzing tabular data.

---

##  Exploring Your DataFrame

- `df.head()`  View the first few rows  
- `df.tail()`  View the last few rows  
- `df.shape`  Get the number of rows and columns  
- `df.columns`  List all column names  

---

##  Cleaning & Handling Missing Data

- `df.isna()` or `df.isnull()`  Identify missing values  
- `df.dropna()`  Remove rows with missing data  
- `df.fillna(value)`  Replace missing values with a default  

---

##  Accessing Data

- `df['column']`  Access a single column  
- `df[['col1', 'col2']]`  Access multiple columns  
- `df.loc[index, 'column']`  Access a cell by label  
- `df.iloc[row_index, col_index]`  Access a cell by position  

---

##  Analyzing Values

- `df.max()` / `df.min()`  Find highest and lowest values  
- `df.idxmax()` / `df.idxmin()`  Find index of those values  

---

##  Sorting & Modifying

- `df.sort_values(by='column')`  Sort by a column  
- `df.insert(loc, column_name, values)`  Add a new column  

---

##  Grouping & Aggregation

- `df.groupby('category')`  Group data by a column  
- Combine with `.mean()`, `.count()`, `.sum()` etc. to summarize  
- Use `select_dtypes(include='number')` to isolate numeric columns before aggregation  

---
